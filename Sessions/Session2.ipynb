```python
#Use CVRAPI () to find and extract all available data for the companies listed in the source file. 
#Store the extracted data in a file and a pandas data frame in your Python application

def cvrapi(cvr):
  request_a = request.Request(
    url='https://cvrapi.dk/api?country=dk&vat=%s' % cvr,
    headers={
      'User-Agent': 'ds-assignment'})
  with contextlib.closing(request.urlopen(request_a)) as response:
    return json.loads(response.read())


#print(data)
```
